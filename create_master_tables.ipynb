```python
#Read paths
import pandas as pd
root_path           = 'C:/Users/ntste/Documents/Insight/Tree Data/'
address_info_path   = root_path + 'LA_County_Parcels_small.csv'
footprint_info_path = root_path + 'Building_Footprints-shp/footprint_nodes.csv'
parcel_info_path    = root_path + 'parcel_locations.csv'
zip_codes_path      = root_path + 'la_zip_codes.txt'
master_file_path    = root_path + 'Master Tables/'
```

```python
#parse large CSV and save parcel outlines to csv
fields = ['WKT','AIN']
parcel_info = footprint_info = pd.read_csv(root_path+'Landbase_Lines___Parcel_Outline-shp/parcels_wkt_2.csv',usecols=fields)
parcel_info.columns = ['PARCEL_WKT','AIN']
parcel_info.to_csv(root_path + 'parcel_locations.csv',index=False)
```

```python
#Read csv files with address, footprint, and parcel info
import pandas as pd
address_info   = pd.read_csv(address_info_path)
footprint_info = pd.read_csv(footprint_info_path)
parcel_info    = pd.read_csv(parcel_info_path,dtype={'PARCEL_WKT': str, 'AIN': str})
```

```python
#Format parcel info
parcel_info = parcel_info.dropna()
parcel_info = parcel_info[(parcel_info.AIN != 'nan')] #remove nan strings that aren't actual NaNs
parcel_info = parcel_info[(parcel_info.AIN != ' ')]
parcel_info = parcel_info[parcel_info.AIN.str.isnumeric()]
parcel_info['AIN'] = parcel_info['AIN'].astype('int64')
```

```python
#perform inner join of all tables by AIN
master_table = address_info.merge(footprint_info, how = 'inner', on = ['AIN'])
master_table = master_table.merge(parcel_info, how = 'inner', on = ['AIN'])
print(master_table.columns)
master_table.head()
```

```python
#read zip codes into list
with open(zip_codes_path, 'r') as f:
    zip_codes = f.readlines()
zip_codes = [i.strip() for i in zip_codes]
print(zip_codes)
```

```python
for code in zip_codes:
    print(code)
    subset_table = master_table.loc[master_table['SitusFullAddress'].str.contains(code, na=False)]
    subset_table.drop(['APN', 'SitusCity','TaxRateCity','ShapeSTArea', 'ShapeSTLength', 'Shape','OBJECTID_y','CODE', 'BLD_ID', 'HEIGHT', 'ELEV', 'AREA', 'LARIAC_SOU', 'LARIAC_DAT','STATUS', 'CODE_NUM'], axis=1)
    subset_table.to_csv(master_file_path + code + '.csv', index = False)
```
